```python
import numpy as np
from scipy.optimize import root_scalar
from consts_and_funcs import *

# 常 量
DRONE_SPEED = 120
T_DEPLOY = 1.5
DELAY = 3.6

 # 无 人 机 速 度 （朝 向FAKE_TARGET， 高 度 保 持 不 变）
drone_horiz = np.array([FY1_INIT[0], FY1_INIT[1]])
drone_horiz_dist = np.linalg.norm(drone_horiz)
drone_vel = -drone_horiz / drone_horiz_dist *DRONE_SPEED
drone_vel = np.array([drone_vel[0], drone_vel[1], 0])

# 投 放 和 起 爆 位 置
deploy_pos = FY1_INIT + drone_vel *T_DEPLOY
explode_pos = deploy_pos + drone_vel *DELAY + np.array([0, 0,-0.5 *G *DELAY**2])
explode_t = T_DEPLOY + DELAY

# 云 团 中 心 函 数
def cloud_center(t):
tau = t-explode_t
if tau < 0:
return explode_pos + np.array([0, 0, -ESCAPE_VEL *
tau
])
if tau > DURATION:
return explode_pos + np.array([0, 0, -SINK_SPEED *
DURATION]) + np.array([np.inf, np.inf, np.inf])
else:
return explode_pos + np.array([0, 0, -SINK_SPEED *
tau
])

if __name__ == '__main__ ':

# 计 算 遮 蔽 时 间
calculate_shield_time = get_calculate_shield_time(
cloud_center , explode_t, 1)
shield_time_inner = calculate_shield_time(INNER_RADIUS)
shield_time_outer = calculate_shield_time(OUTER_RADIUS)
shield_time_average = (shield_time_inner +
shield_time_outer) / 2

# 输 出 结 果
print(f"有 效 遮 蔽 时 长 (内 切 球， 半 径 {INNER_RADIUS}m): {shield_time_inner:.6f} s")
print(f"有 效 遮 蔽 时 长 (外 接 球， 半 径 {OUTER_RADIUS:.3f}m): {shield_time_outer:.6f} s")
print(f"内 切 球 和 外 接 球 有 效 遮 蔽 时 长 的 平 均 值: {shield_time_average:.6f} s")
```
